```python
# Initialize the grid
grid = [[" " for _ in range(3)] for _ in range(3)]

# Function to print the grid
def print_grid():
    for row in grid:
        print("|" + "|".join(row) + "|")

# Function to handle player turns
def player_turn(turn_player1):
    print(f"It's {player2 if turn_player1 else player1}'s turn")
    return not turn_player1

# Function to write to a cell
def write_cell(cell, player_symbol):
    grid[(cell - 1) // 3][(cell - 1) % 3] = player_symbol

# Function to check if a cell is free
def free_cell(cell):
    return grid[(cell - 1) // 3][(cell - 1) % 3] == " "

# Function to check for a win
def win_check():
    # Check rows, columns, and diagonals
    lines = [*grid, *zip(*grid), [grid[i][i] for i in range(3)], [grid[i][2 - i] for i in range(3)]]
    for line in lines:
        if all(cell == player1_symbol for cell in line):
            return False, player1
        if all(cell == player2_symbol for cell in line):
            return False, player2
    return True, "" if any(" " in row for row in grid) else (False, "")

# Game opening
print("Welcome to Tic-Tac-Toe!")
print_grid()
player1 = input("Please enter the name of player 1: ")
player1_symbol = input("Please enter the symbol of player 1: ")
player2 = input("Please enter the name of player 2: ")
player2_symbol = input("Please enter the symbol of player 2: ")

game, turn_player1 = True, True

# Game loop
while game:
    print_grid()
    turn_player1 = player_turn(turn_player1)
    player_symbol = player1_symbol if not turn_player1 else player2_symbol

    while True:
        cell = int(input("Please enter a number for your cell (1 to 9): "))
        if free_cell(cell):
            write_cell(cell, player_symbol)
            break
        print("This cell is not free. Please choose another.")

    game, winner = win_check()

# End of game
print(f"Winner is {winner}!" if winner else "Grid is full: it's a draw!")
```
